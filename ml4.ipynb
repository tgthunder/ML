```python
x=2
lr=0.001
max_iter=10000
iter=0
precision=0.000001
prev_step_size=1
gf=lambda x:(x+3)**2
```

```python
f=[]
while(precision<prev_step_size and iter<max_iter):
    prev=x
    x=x-lr*gf(prev)
    prev_step_size=abs(x-prev)
    iter+=1
    print("iteration :",iter,"value :",x)
    f.append(x)
```

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(f)
```
